```python
def upto(value):
  if "Expandable Upto" in value:
    part=value.split("Expandable Upto",1)
    capacity=part[1].strip()
    return capacity
  else:
    return None
```

```python
df["expandable"]=df["Descriptions"].apply(upto).str.split().str[0]
```

```python
df.head()
```

```python
def upto1(value):
  if "cm" in value:
    part=value.split("cm (")
    con=part[1].strip().split(" ",1)[0]
    return con
  else:
    None
```

```python
df["Descriptions"][1]
```

```python
df["display_size(inches)"]=df["Descriptions"].apply(upto1)
```

```python
def camera(value):
  if "Display" in value:
    part=value.split("Display")
    con=part[1].strip()
    return con
  else:
    None
```

```python
df["Descriptions"].str.split("Mp")[0]
```

```python
df["Descriptions"][1]
```

```python
df["Descriptions"].apply(camera)
```

```python
def camerasplit(value):
  if "|" in value:
    newpart=value.split("|")
    con=newpart[0].strip()
    con=con.replace("MP",'')
    new=con.split("Mp")
    new_con=new[0].strip()
    return new_con
  else:
    if "Rear" in value:
      new=value.split("Rear")
      con1=new[0].strip()
      con1=con1.replace("MP",'')
      return con1
```

```python
df["camera"]=df["Descriptions"].apply(camera).apply(camerasplit)
```

```python
df.rename(columns={"camera":"rear_camera"},inplace=True)
```
